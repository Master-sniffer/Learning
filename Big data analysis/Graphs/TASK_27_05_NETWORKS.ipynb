```python
class Codes:

  def __init__(self):
    print("hey")

  def avtomat1(self):
    return self.avto1
  
  def avtomat2(self):
    return self.avto2
  
  def edggee(self):
    return self.graph_edges
  
  def graphAndPos(self, graph,positions ):
    self.graph=graph
    self.positions=positions
  
  def returngraphsandpos(self):
    return self.graph, self.positions
  
  def iters(self):
    for i in self.graph_nodes:
      if i in self.avto1:
        self.graph.nodes[i]['color'] = 'purple'
      else:
        self.graph.nodes[i]['color'] = 'orange'
    for i in self.graph_edges:
      self.graph[i[0]][i[1]]['p'] = round(random.uniform(0.4,0.7), 2)

  def draw_pic(self):
    self.graph_edges = graph.edges()
    self.graph_nodes = graph.nodes()
    self.avto1 = random.sample(list(self.graph_nodes), k=random.randint(1,50))
    self.avto2 = [i for i in self.graph_nodes if not i in self.avto1]
    self.iters()
    plt.figure(figsize = (20, 20))
    nx.draw_networkx_nodes(self.graph, self.positions, nodelist=self.avto1, node_color="tab:purple")
    nx.draw_networkx_nodes(self.graph, self.positions, nodelist=self.avto2, node_color="tab:orange")
    nx.draw_networkx_labels(self.graph,self.positions)
    self.labels = nx.get_edge_attributes(self.graph, "p")
    nx.draw_networkx_edge_labels(self.graph, pos=self.positions, edge_labels=self.labels)
    nx.draw_networkx_edges(self.graph, pos=self.positions, edgelist=self.graph_edges)
    plt.show()
```
