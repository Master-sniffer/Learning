2.1 На основе файла `contributors_sample.json` создайте словарь следующего вида: 
```
{
    должность: [список username людей, занимавших эту должность]
}
```

````python
import pandas as pd
import json
from bs4 import BeautifulSoup
import pickle


with open ('/content/drive/MyDrive/Colab Notebooks/lab 4/contributors_sample.json', 'r', encoding='utf-8') as fp:
  data = json.load(fp)

#output = set()
d = {}
for i in data:
  for j in i['jobs']:
    if j not in d:
      d[j]=[]
    
    if i['username'] not in d[j]:
      #print (d[j])
      d[j].append(i['username']) 

with open('mypickle.pickle', 'wb') as f:
    pickle.dump(d, f)

with open("sample.json", "w") as outfile:
    json.dump(d, outfile, indent=4)

````
