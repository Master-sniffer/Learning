5.2 При помощи объединения таблиц и группировок, создайте `DataFrame`, состоящий из трех столбцов: `recipe_id`, `name`, `review_count`. У рецептов, для которых отсутствуют отзывы, в соответствущем столбце должен быть указан 0. Подтвердите правильность работы вашего кода, выбрав рецепт, не имеющий отзывов, и выведя на экран строку из полученного `DataFrame`, содержащую информацию об этом отзыве.


```python
reviews = pd.read_csv('/content/reviews_sample.csv', sep =',', names=["id","user_id","recipe_id","date","rating","review"], header=None , skiprows=1)
recipes  = pd.read_csv('/content/recipes_sample.csv', sep =',', names=["name","id","minutes","user_id","submitted","n_steps","description","n_ingredients"], header=None, skiprows=1)
recipes=recipes.join( reviews.set_index('recipe_id')[['review']], on='id')
recipes.pop("n_steps")
recipes.pop("minutes")
recipes.pop("submitted")
recipes.pop("description")
recipes.pop("user_id")
recipes.pop("n_ingredients")
#recipes.groupby('id').review.apply(lambda x: x.count()-x.isna().sum())
recipes.review.notnull().groupby([recipes['name'],recipes['id']]).sum().astype(int).reset_index(name='count')
```

### Сохранение таблиц `pd.DataFrame`

6.2 Воспользовавшись `pd.ExcelWriter`, cохраните результаты 5.1 и 5.2 в файл: на лист с названием `Рецепты с оценками` сохраните результаты выполнения 5.1; на лист с названием `Количество отзывов по рецептам` сохраните результаты выполнения 5.2.

```python
reviews = pd.read_csv('/content/reviews_sample.csv', sep =',', names=["id","user_id","recipe_id","date","rating","review"], header=None , skiprows=1)
recipes  = pd.read_csv('/content/recipes_sample.csv', sep =',', names=["name","id","minutes","user_id","submitted","n_steps","description","n_ingredients"], header=None, skiprows=1)
reviews = reviews[reviews['review'].notna()]
reviews.pop("recipe_id")
reviews.pop("date")
reviews.pop("review")
res1=reviews.join( recipes.set_index('id')[['name']], on='user_id')

with pd.ExcelWriter("Рецепты с оценками.xlsx") as writer:
  res1.to_excel(writer)

reviews = pd.read_csv('/content/reviews_sample.csv', sep =',', names=["id","user_id","recipe_id","date","rating","review"], header=None , skiprows=1)
recipes  = pd.read_csv('/content/recipes_sample.csv', sep =',', names=["name","id","minutes","user_id","submitted","n_steps","description","n_ingredients"], header=None, skiprows=1)
recipes=recipes.join( reviews.set_index('recipe_id')[['review']], on='id')
recipes.pop("n_steps")
recipes.pop("minutes")
recipes.pop("submitted")
recipes.pop("description")
recipes.pop("user_id")
recipes.pop("n_ingredients")
#recipes.groupby('id').review.apply(lambda x: x.count()-x.isna().sum())
res2=recipes.review.notnull().groupby([recipes['name'],recipes['id']]).sum().astype(int).reset_index(name='count')

with pd.ExcelWriter("Количество отзывов по рецептам.xlsx") as writer:
  res2.to_excel(writer)
```
