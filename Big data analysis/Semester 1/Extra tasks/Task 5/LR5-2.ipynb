```python
d=[]
# получаем страницы при помощи цикла
for j in range(10):
    #указываем url и get параметры запроса
    url = 'https://www.avito.ru/moskva/tovary_dlya_detey_i_igrushki?cd=1'
    # указываем get параметр с помощью которого определяется номер страницы
    par = {'p': j}
    # записываем ответ сервера в переменную r
    r = requests.get(url, params=par)
    
    # получаем объект  BeautifulSoup и записываем в переменную soup
    soup = BeautifulSoup(r.text, 'html.parser')
    # с помощью циклам перебераем товары на странице и получаем из них нужные параметры
    for i in range(40):
           # получаем название товара
           item = soup.find_all(class_="item-description-title-link")[i].get_text().replace("\n","")
           # получаем цену товара
           price = soup.find_all(class_="price")[i].get_text().replace("\n","")
           price = price.replace(" ", "")
           #получаем ссылку на товар
           link =url+ soup.find_all(class_="item-description-title-link")[i]['href']
           
           # добавляем данные о товаре в список
           d.append([item, price, link])
 

```

```python
d
```

```python
import pandas as pd
df=pd.DataFrame(d,columns=['название',"цена","ссылка"])
df
```

```python
df.to_csv("Avito.csv",sep=";",index=0)
```

```python
base_url = 'https://www.nhl.com/kings/roster'
r = requests.get(base_url)
 
soup = BeautifulSoup(r.content, 'html.parser')
```

```python
len(soup.find_all(class_='dropdown-menu dropdown-menu--even-width')[1].find_all('a'))
```

```python
l=[]
num=len(soup.find_all(class_='dropdown-menu dropdown-menu--even-width')[1].find_all('a'))
for i in range(num):
    link=soup.find_all(class_='dropdown-menu dropdown-menu--even-width')[1].find_all('a')[i]['href']
    l.append(link)
```

```python
d=[]
for j in range(len(l)):

    url = 'https://www.nhl.com'+l[j]
    r = requests.get(url)
        # получаем объект  BeautifulSoup и записываем в переменную soup
    soup = BeautifulSoup(r.text, 'html.parser')
    divs = soup.find_all('div', attrs={'class': 'name-col__list'})
    for i in range(len(divs)):
        firstName = soup.find_all(class_="name-col__item name-col__firstName")[i].get_text().replace("\n","")
        lastName = soup.find_all(class_="name-col__item name-col__lastName")[i].get_text().replace("\n","")
               # получаем имя, фамилию игрока
        num = soup.find_all(class_="number-col fixed-width-font")[i].get_text().replace("\n","")
               # получаем номер
        pos =soup.find_all(class_="position-col fixed-width-font")[i].get_text().replace("\n","")
            # получаем позицию игрока
        weight = round(int(soup.find_all(class_="weight-col fixed-width-font")[i].get_text().replace("\n",""))*0.45)
               # получаем вес игрока
        bd = soup.find_all(class_="all-but-xs-sm-md")[2*i+1].get_text().replace("\n","")
         # получаем дату рождения игрока      
               # добавляем данные о игроке в список
        d.append([firstName, lastName, num,pos,weight,bd])
```

```python
d
```

```python
import pandas as pd
df=pd.DataFrame(d,columns=['firstName', 'lastName', 'num','pos','weight','bd'])
df
```

```python
df.to_csv("hockey.csv",sep=";",index=0)
```
